```python
def sense(x):
    return x

def simulate(Δt, x, u):
    x += Δt * u
    return x

def control(t, y):
    ux = -sin(t)
    uy = cos(t)
    uz = 1
    return array([ux, uy, uz])

tf = 6.*pi
Δt = 0.01*pi
time = linspace(0,tf, int(tf / Δt) + 1)

x = array([0., 0., 0.])
x_log = [copy(x)]

for t in time:
    y = sense(x)
    u = control(t, y)
    x = simulate(Δt, x, u)
    x_log.append(copy(x))
    
x_log = array(x_log)
```

```python
fig = figure()
ax = axes(projection='3d')
ax.plot3D(x_log[:,0], x_log[:,1], x_log[:,2])
```
